```python
import pandas as pd
```

```python
# import zillow home price data
zillowdata = pd.read_csv('C:/Users/chris/Desktop/GitHub/zillow.csv')
```

```python
# Set no max rows and columns to view data
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# Reduce and clean data for analysis

# 1Reduce data to only North Carolina
# 2The most recent month home price 
```

```python
# Reduce the dataset to only North Carolina
zillow1 = zillowdata[zillowdata['State'].str.contains("NC")]
```

```python
# Keep only data information and most current home prices
zillow2 = zillow1[['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName',
       'State', 'City', 'Metro', 'CountyName', '2023-02-28']]
```

```python
# Drop columns
zillow2.drop(['RegionID', 'RegionType', 'StateName'], axis = 1, inplace = True)
```

```python
# Rename zip code column
zillow2.rename(columns={'RegionName':'Zip'})
```

```python
zillow2.rename(columns={'2023-02-28':'AveragePrice'}, inplace = True)
```

```python
# Change the home price from being displayed in scientific notation
pd.options.display.float_format = '{:.2f}'.format
```

```python
# create new dataframe to work with
zillow3 = zillow2
```

```python
# Create a new column showing 20% downpayment
zillow3['Down_payment_20%'] = zillow3['AveragePrice'].apply(lambda x: x * .2)
```

```python
# Add estimated mortgage based on 7%
zillow4 = zillow3
```
